# Iniciar o repositório

## Lista de comandos do Git
* Link: https://git-scm.com/docs/git#_git_commands

## Criando o repositório

> git init

## Checar status do repositório


> git status


## Commitar (enviar)

### OBS: Commit não envia para o servidor, simplesmente diz que está pronto para ser salvo!

> git commit -m `"mensagem_aqui"`
* -m: flag de mensagem

## Criando a branch (ramo)

> git branch -M main
 * branch principal, não precisa se chamar main/master, mas é uma convenção

## Sincronizando com o repositório na nuvem

> git remote add origin `link HTTP do repositório`

## Enviando para a branch especificada

> git push -u origin `nome da branch`

* -u: flag para indicar o repositório principal

## Mudar o remote origin (repositório na nuvem)

#### Para verificar qual o repositório

> git remote -v
* -v: flag para indicar detalhamento (verbose)

#### Para remover o repositório

> git remote rm origin

# Adicionar arquivos no versionamento


* > git add `nome_arquivo`

* > git add * (para adicionar todos os arquivos como untracked no diretório atual)
* > git add . (para adicionar todos os arquivos e subdiretórios)

## Adicionar os novos arquivos na fila (commit)

> git commit -a -m 
* -a: flag para adicionar vários arquivos
* -m: flag para adicionar mensagem


## Enviando para o repositório

> git push

# Receber alterações

> git pull

## Clonando repositório

> git clone `link`

# Remover arquivos do repositório

> git rm `nome do arquivo`

> Após ser retirado o arquivo, devemos fazer commit para atualizar o repositório

* > git commit -a -m "`mensagem`"
* > git push

# Mover ou renomear arquivos

## Mover o arquivo


* > git mv `nome arquivo` `pasta/nome arquivo`
* Ex.: git mv teste.txt repositorio_teste/teste.txt


## Renomear arquivo

#### caso esteja em outra pasta
* > git mv `pasta/nome arquivo` `pasta/novo nome`

## Após as alterações devemos fazer commit!

* > git commit -a -m "`mensagem`"
* > git push

# Ignorar arquivos

* Devemos inserir um arquivo chamado ``.gitignore`` na raiz do projeto
* Nele podemos inserir todos os arquivos de não devem entrar no versionamento
* isso é útil para arquivos gerados automaticamente ou arquivos que contêm ``informações sensíveis``

* > git add .
* > git commit -a -m "`mensagem`"
* > git push 

<div class="alert alert-block alert-warning">
Caso tenha criado um arquivo .gitignore após a criação de algum arquivo que queira ignorar use <b>este comando</b>!

* > git rm -r --cached .

* > git rm : remove os arquivos <br>
* > -r : flag (recursive) todos os diretórios e subdiretórios serão adicionados na remoção<br>
* > --cached: Remove apenas no índice (área de staging), não os deletando do disco<br>
* > '.' : Se refere ao diretório atual e subdiretórios<br>

* > git add .<br>
* > git commit -m `mensagem`

# Desfazer todas as alterações

<div class="alert alert-block alert-warning">
