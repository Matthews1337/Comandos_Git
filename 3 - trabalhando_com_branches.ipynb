# O que é uma branch?

* Branch é a forma que o git separa as versões dos projetos
* Quando um projeto é criado ele inicia na branch master
* Geralmente cada nova feature de um projeto fica em um branch separado
* Após a finalização das alterações os branchs são unidos para ter o código-fonte final
* Normalmente em desenvolvimento de software cada branch possui uma nova aplicação em desenvolvimento
* Geralmente se cria uma nova funcionalidade em cima da branch que está em produção, ou seja , clona o repositório, cria uma nova branch, desenvolve a funcionalidade e salva a funcionalidade na nova branch

# Criar e visualizar branches

* > Para visualizar as branches disponíveis: `git branch`
* > Para criar uma branch : `git branch nome`

---

# Deletar branches

#### Geralmente se usa o delete quando a branch foi criada errada

* Podemos deletar uma branch com a flag ``-d`` ou ``--delete``
* > git branch -d nome_branch


<div class="alert alert-block alert-warning">
<b>CUIDADO!</b> Uma vez apagada uma branch, não tem como recuperá-la!


---

# Mudar de branch

#### Ao criar, fazer alguma alteração no projeto e fazer commit, se fizermos `git checkout main` podemos mudar para outra branch

<div class="alert alert-block alert-warning">
<b>CUIDADO!</b> Sempre salvar as alterações (commit) antes de mudar de branch, caso contrário as alterações podem ser transferidas para outra branch


* > git checkout `nome_branch` 
* git checkout -b ``nome_branch`` para criar e mudar automaticamente para a nova branch

---

# Unir branches

* Normalmente é por meio deste comando que se recebe atualizações de outros devs
* Caso a branch que se está trabalhando estiver desatualizada comparada com a main/master, devemos fazer um `git merge main` para atualizar com as mais novas funcionalidades

* > git merge `nome_branch`

Após o merge devemos fazer o push

# Stash (lixo)

* Podemos salvar as modificações atuais para prosseguir com uma outra abordagem de solução e não perder o código
* Após o comando a branch será resetado para a sua versão de acordo com o repo

* > git stash

## Recuperar arquivos do stash

* > Podemos verificar as stash's criadas pelo comando `git stash list`
* > podemos recuperar a stash com o comando `git stash nome_arquivo`
* > Para recuperar algum arquivo da stash: `git apply indice_stash`
* > Para visualizar as alterações: `git stash show -p indice_stash`

## Remover stash

* > Para limpar totalmente a stash de uma branch podemos utilizar o comando `git stash clear`
* > Caso seja necessário deletar uma stash específica podemos utilizar: `git stash drop nome_stash`

# Tags

* > Tags funcionam como "checkpoints" de uma branch. São utilizadas para demarcar estágios do desenvolvimento de algum recurso. Praticamente uma branch de uma branch

> git tag -a ``nome_tag`` -m `"mensagem"`

* para ver as tags: `git tag`


### Verificar e alterar tags

* > Podemos verificar uma tag com o comando `git show nome_tag`
* > Podemos trocar de tags com o comando `git checkout nome_tag`

## Enviar e compartilhar tags

* > As tags podem ser enviadas para o repositório de código, sendo compartilhada entre os devs
* > O comando é git push origin `nome_repo`
* > Caso quiser enviar mais tags: git push origin --tags
