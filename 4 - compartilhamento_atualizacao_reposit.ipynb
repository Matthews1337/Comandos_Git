# Receber alterações

* 1 - Após criar uma nova branch com uma nova funcionalidade, devemos commitar as alterações e fazer o push para o repositório.
* 2 - A pessoa responsável pelo merge deverá ir na branch main, fazer o pull para receber as novas funcionalidades do outro dev e fazer o merge com as branches novas.
