# Utilizando o Remote

 * > Com o git remote podemos fazer algumas ações como: Adicionar um repo para trackear ou remover
 * > Quando criamos um repo remoto, adicionamos ele ao git com:
 * > git remote add origin `URL`

Verificar os repositórios: <br>
* > git remote -v
<br>

Remover o caminho do repositório: <br>
* > git remote rm origin
<br>

Adicionar novo caminho do repositório: <br>
* > git remote add origin `URL`


# Submódulos

Submódulos são repositórios separados, mas funcionam como um repositório dentro de outro repositório, estes são separados por pastas quando criado o submódulo!

* > git submodule add `URL` 

Primeiro deve-se criar um novo repositório:



``Depois de criado o novo repositório, devemos pegar o link novo repositório e adicionar como um submódulo ao projeto atual!``


# Atualizar submódulos

ATENÇÃO: Para fazer/receber qualquer atualização no submódulo, certifique-se que está dentro da pasta do submódulo!

* > git push --recurse-submodules=on-demand
